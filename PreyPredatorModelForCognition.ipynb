```python
%matplotlib inline 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
# 2d Linear system
def LotkaVolterra(state,t):
    x = state[0]
    y = state[1]
     
    # Vanderpol as an example
    a = 0.1
    b = 0.1
    c = 0.1
    d = 0.2
    
    # Main equation here
    dx = x*(a - b*y)
    dy = -y*(d - c*x)
    
    # Return the state derivatives
    return [dx,dy]
```

```python
t = np.arange(0, 500, 1) # Times

state0 = [0.5,0.5]
state = odeint(LotkaVolterra,state0,t)
plt.figure()
plt.plot(t,state)
plt.ylim([0,8])
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.legend(('x (prey)','y (predator)'))
plt.title('Lotka-Volterra equations')
```
